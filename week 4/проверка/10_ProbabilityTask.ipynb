```python
#импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Логонормальное распределение

```python
#сгенерируем логонормальное распределение
Myu = 1.0
Sigma2 = 1.0
lognorm_rv = sts.lognorm(scale = np.exp(Myu),s = np.sqrt(Sigma2))
```

```python
#создадим выборку размером 1000
smp = lognorm_rv.rvs(1000)
#smp
```

```python
#построим гистограмму выборки
plt.hist(smp, density=True)

#сгенерируем значения x для отрисовки теоретической плотности распределения
x = np.linspace(0,15,1000)
#найдем значения функции плотности вероятности
pdf = lognorm_rv.pdf(x)
#добавим теоретическую плотность на график
plt.plot(x, pdf, label='theoretical pdf')
#подпишем оси
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

### Создание выборочных средних

```python
#создаем массив для записи выборочных средних
vs = []
for i in range(1000):
    #добавим новый элемент в массив
    vs.append([])
    #построим выборки различных размеров
    smp5 = lognorm_rv.rvs(5)
    smp10 = lognorm_rv.rvs(10)
    smp50 = lognorm_rv.rvs(50)
    #запишем выборочные средние в массив
    vs[i]=(smp5.mean(),smp10.mean(),smp50.mean())
```

```python
#создадим массив numPy, т.к. из него можно выбирать столбцы
npvs = np.array(vs)
```

```python
#построим гистограмму распределения выборочных средних для n=5
plt.hist(npvs[:,0], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=5')
```

```python
#построим гистограмму распределения выборочных средних для n=10
plt.hist(npvs[:,1], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=10')
```

```python
#построим гистограмму распределения выборочных средних для n=50
plt.hist(npvs[:,2], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=50')
```

### Расчет значений параметров нормальных распределений
Необходимо построить график нормального распределения, а для этого требуется расчитать матожидание и дисперсию

```python
Var = (np.exp(Sigma2)-1) * np.exp(2*Myu+Sigma2)
E = np.exp(Myu+Sigma2/2)
print Var
print E

#Матожидание одинаково для всех n

#Расчитаем стандартное отклонение для последующей отрисовки графиков
#для n=5
std5=np.sqrt(Var/5)
#для n=10
std10=np.sqrt(Var/10)
#для n=50
std50=np.sqrt(Var/50)
```

```python
#для n=5

#зададим нормальное распределение с расчитанными ранее параметрами
norm_rv5 = sts.norm(E, std5)

plt.hist(npvs[:,0], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=5')

#зададим точки для отрисовки нормального распределения
x5=np.linspace(1,10,1000)
pdf5 = norm_rv5.pdf(x5)
plt.plot(x5, pdf5, label='theoretical pdf')
```

```python
#для n=10

#зададим нормальное распределение с расчитанными ранее параметрами
norm_rv10 = sts.norm(E, std10)

plt.hist(npvs[:,1], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=10')

#зададим точки для отрисовки нормального распределения
x10=np.linspace(0,10,1000)
pdf10 = norm_rv10.pdf(x10)
plt.plot(x10, pdf10, label='theoretical pdf')
```

```python
#для n=50

#зададим нормальное распределение с расчитанными ранее параметрами
norm_rv50 = sts.norm(E, std50)

plt.hist(npvs[:,2], density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x-avg, n=50')

#зададим точки для отрисовки нормального распределения
x50=np.linspace(0,10,1000)
pdf50 = norm_rv50.pdf(x50)
plt.plot(x50, pdf50, label='theoretical pdf')
```

### Выводы по результатам

Разница между полученными распределениями заключается в том, что при росте n распределения начинают все больше походить на нормальное распределение.
Также с ростом n увеличивается точность аппроксимации распределения выборочных средних.
