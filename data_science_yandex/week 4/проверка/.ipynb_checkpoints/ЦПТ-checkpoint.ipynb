```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Для начала, сгенерируем выборку из случаев выпадания какой либо стороны для игральной косточки. В данном примере я взял 100 подрасываний косточки

```python
sample = np.random.choice([1,2,3,4,5,6], 1000)
```

```python
from collections import Counter

c = Counter(sample)#Выпадание каждой и з сторон

print("Число выпадений каждой из сторон:")    
print(c)

# Вероятность выпадения каждой из сторон
print("Вероятность выпадения каждой из сторон косточки:")
print({k: v/1000 for k, v in c.items()})
```

### Выборка из 100 случаев из стандартного нормального распределения (с $\mu=0$ и $\sigma^2=1$):

```python
norm_rv = sts.norm(0, 1)
sample = norm_rv.rvs(1000)
```

Получаем график нормельного распределения, для более точного вывода количество bins увеличим до 12

```python
plt.hist(sample,bins=12, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

Используем ракетную функцию для сглаживания распределения. Для сравнения возьмем KDE и теоретическое распределение и видим чем больше выборка тем более вероятно они похожи друг на друга

```python
df = pd.DataFrame(sample, columns=['KDE'])
ax = df.plot(kind='density')

x = np.linspace(-4,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
