```python
import numpy as np
import math
import scipy.linalg as scipylg
import matplotlib.pyplot as plt 

def myfun(c):
    y = math.sin(c/5)* math.exp(c/10)+ 5*math.exp(-c/2)
    return y
a1=np.array([[1, 1],[1, 15]])
x1=np.array([1, 15])
b1=([myfun(x1[0]), myfun(x1[1])])

#b.append(myfun(1))
#b.append(myfun(15))

x_0=scipylg.solve(a1,b1)
print(x_0)


#y0 = math.sin(x1/5)* math.exp(x1/10)+ 5*math.exp(-x1/2)
#plt.plot(x1,b1)
y1=x_0[0]*1+x_0[1]*1
#plt.plot(x1,myfun(x1))
plt.plot(x1,y1,'r--')
```

```python
a2=np.array([[1, 1, 1],[1, 8, 64],[1, 15, 225]])
x2=np.array([1, 8, 15])
b2=([myfun(x2[0]), myfun(x2[1]),myfun(x2[2])])
x_1=scipylg.solve(a2,b2)
print(x_1)

plt.plot(x2,b2)
plt.plot(x_1,b2,'r--')
```
